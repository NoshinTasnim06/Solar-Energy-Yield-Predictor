```python
plantgen = pd.read_csv('/Users/noshintasnim/Downloads/Solar-Energy-Yield-Predictor/Raw Data/Plant_Generation_&_Weather_censor_Data/Plant_2_Generation_Data.csv')
plantweather = pd.read_csv('/Users/noshintasnim/Downloads/Solar-Energy-Yield-Predictor/Raw Data/Plant_Generation_&_Weather_censor_Data/Plant_2_Weather_Sensor_Data.csv')
```

```python
plantgen['DATE_TIME'] = pd.to_datetime(plantgen['DATE_TIME'], errors='coerce')
plantweather['DATE_TIME'] = pd.to_datetime(plantweather['DATE_TIME'], errors='coerce')

plantgen_agg = plantgen.groupby('DATE_TIME').agg({'PLANT_ID': 'first', 'DC_POWER': 'sum', 
                    'AC_POWER': 'sum', 'DAILY_YIELD': 'sum'}).reset_index()
plantgen_agg.head()
```

```python
plantweather.drop(columns=['SOURCE_KEY'], inplace=True)
plantweather.head()
```

```python
plantdata = pd.merge(plantgen_agg, plantweather, on=['DATE_TIME', 'PLANT_ID'], how='left')
plantdata.head()
```

```python
plantdata.insert(3, 'DCE', (plantdata['DC_POWER'] / plantdata['DC_POWER'].max()).fillna(0))
plantdata.insert(5, 'ACE', (plantdata['AC_POWER'] / plantdata['AC_POWER'].max()).fillna(0))
plantdata.insert(6, 'CE', (plantdata['AC_POWER'] / plantdata['DC_POWER'].max()).fillna(0))
```

```python
plantdata.head(10)
```

```python
plantdata = plantdata.rename(columns={'DCE': 'DC_EFFICIENCY', 'ACE': 'AC_EFFICIENCY', 'CE': 'CONVERSION_EFFICIENCY'})
plantdata.head(10)
```

```python
plantdata.to_csv('SOLAR_PLANT_DATA(GENERATION_AND_WEATHER).csv', index=False)
```
