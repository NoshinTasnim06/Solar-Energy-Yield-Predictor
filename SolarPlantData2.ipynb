```python
gd1['DATE_TIME'] = pd.to_datetime(gd1['DATE_TIME'], errors='coerce')
gd2['DATE_TIME'] = pd.to_datetime(gd2['DATE_TIME'], errors='coerce')
wsd1['DATE_TIME'] = pd.to_datetime(wsd1['DATE_TIME'], errors='coerce')
wsd2['DATE_TIME'] = pd.to_datetime(wsd2['DATE_TIME'], errors='coerce')
```

```python
gd1.head(10)
```

```python
gd1_agg = gd1.groupby('DATE_TIME').agg({'PLANT_ID': 'first', 'DC_POWER': 'sum', 'AC_POWER': 'sum', 'DAILY_YIELD': 'sum'}).reset_index()
```

```python
gd1_agg.iloc[55:75]
```

```python
gd2_agg = gd2.groupby('DATE_TIME').agg({'PLANT_ID': 'first', 'DC_POWER': 'sum', 'AC_POWER': 'sum', 'DAILY_YIELD': 'sum'}).reset_index()
```

```python
gd2_agg.head(10)
```

```python
combined_pl1 = pd.merge(gd1_agg, wsd1, on=['DATE_TIME', 'PLANT_ID'], how='left')
```

```python
combined_pl1.head(10)
```

```python
combined_pl2 = pd.merge(gd2_agg, wsd2, on=['DATE_TIME', 'PLANT_ID'], how='left')
```

```python
combined_pl2.iloc[655:680]
```

```python
spd2 = pd.concat([combined_pl1, combined_pl2], axis=0, ignore_index=True)
```

```python
spd2.iloc[655:680]
```

```python
spd2['SOURCE_KEY'] = spd2['SOURCE_KEY'].astype('string')

spd2 = spd2.rename(columns={'DATE_TIME': 'Timestamp', 'SolarPlantData1.ipynb' 'IRRIDATION': 'Irradiation', 
                            'AMBIENT_TEMPERATURE': 'Ambient_Temperature', 'MODULE_TEMPERATURE': 'Module_Temperature', 
                            'DAILY_YIELD': 'Power_Generation'})

spd2.drop(columns=['SOURCE_KEY', 'DC_POWER', 'AC_POWER'], inplace=True)
```
